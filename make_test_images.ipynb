### Small shifted images to test pixel alignment code
#### Load two images and display

```python
import matplotlib.pyplot as plt                      # import matplotlib
import numpy as np
img1 = plt.imread('MC20210606_2333_01_1933edt.jpg')  # load image into np array
plt.imshow(img1)                                     # display image
```

```python
img2 = plt.imread('MC20210606_2233_01_1833edt.jpg')  # load second image
plt.imshow(img2)                                     # display image
```

#### Crop the first image, display, and save as `small1.jpg`

```python
print("Size of image original:",np.shape(img1))
img1s = img1[100:700,700:1700,:]
print("Size of small image:",np.shape(img1s))
plt.imshow(img1s)
plt.imsave('small1.jpg',img1s)
```

#### Save a cropped, shifted version of the first image as `small1_shifted.jpg`

```python
img1ss = img1[115:715,733:1733,:]
print("Size of shifted small image:",np.shape(img1ss))
plt.imshow(img1ss)
plt.imsave('small1_shifted.jpg',img1ss)
```

#### Save a cropped, shifted version of the second image as `small2_shifted.jpg`

```python
img2ss = img2[110:710,730:1730,:]
print("Size of shifted small image:",np.shape(img2ss))
plt.imshow(img2ss)
plt.imsave('small2_shifted.jpg',img2ss)
```
